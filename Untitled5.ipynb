```python
cols_to_check = ['Indoor Temp','Indoor Hum','Outdoor Temp','Outdoor Hum']

# Replace zeros with NaN (if not done already)
df[cols_to_check] = df[cols_to_check].replace(0, np.nan)

# Now check consecutive NaN gaps
for col in cols_to_check:
    is_nan = df[col].isna()
    gaps = (is_nan != is_nan.shift()).cumsum()  # label consecutive blocks
    gap_sizes = is_nan.groupby(gaps).sum()
    large_gaps = gap_sizes[gap_sizes > 0]  # only gaps > 0
    print(f"{col} gap sizes:\n{large_gaps}\n")
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df = df.set_index('Datetime')

# -------------------------
# 2 Define columns
# -------------------------
cols_to_interpolate = ['Indoor Temp','Indoor Hum']         # indoor, small gaps
cols_to_check_large_gaps = ['Outdoor Temp','Outdoor Hum'] # outdoor, may have large gaps

# -------------------------
# 3 Replace zeros with NaN
# -------------------------
df[cols_to_interpolate + cols_to_check_large_gaps] = df[cols_to_interpolate + cols_to_check_large_gaps].replace(0, np.nan)

# -------------------------
# 4 Handle indoor columns (small gaps)
# -------------------------
df[cols_to_interpolate] = df[cols_to_interpolate].interpolate(method='time')

# -------------------------
# 5 Handle outdoor columns (small + large gaps)
# -------------------------
def mark_large_gaps(series, max_gap=50):
    """
    Returns a Series with the size of consecutive NaN gaps.
    """
    is_nan = series.isna()
    gap_sizes = is_nan.groupby((is_nan != is_nan.shift()).cumsum()).transform('sum')
    return gap_sizes

for col in cols_to_check_large_gaps:
    gap_sizes = mark_large_gaps(df[col])
    # Small gaps: interpolate
    df.loc[gap_sizes <= 50, col] = df.loc[gap_sizes <= 50, col].interpolate(method='time')
    # Large gaps: forward fill (you could also use backward fill if needed)
    df.loc[gap_sizes > 50, col] = df.loc[gap_sizes > 50, col].fillna(method='ffill')

# -------------------------
# 6 Check results
# -------------------------
print(df.isna().sum())
```
