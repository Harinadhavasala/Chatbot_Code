```python
cols_to_check = ['Indoor Temp','Indoor Hum','Outdoor Temp','Outdoor Hum']

# Replace zeros with NaN (if not done already)
df[cols_to_check] = df[cols_to_check].replace(0, np.nan)

# Now check consecutive NaN gaps
for col in cols_to_check:
    is_nan = df[col].isna()
    gaps = (is_nan != is_nan.shift()).cumsum()  # label consecutive blocks
    gap_sizes = is_nan.groupby(gaps).sum()
    large_gaps = gap_sizes[gap_sizes > 0]  # only gaps > 0
    print(f"{col} gap sizes:\n{large_gaps}\n")
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])

# Define phases as (start_date, end_date)
phases = {
    "Phase 2": ("2025-07-22", "2025-08-15"),
    "Phase 3": ("2025-09-01", "2025-09-24"),
    "Phase 4": ("2025-10-22", "2025-11-14"),
}

for name, (start, end) in phases.items():
    mask = (df['Datetime'] >= start) & (df['Datetime'] <= end)
    count = df[mask].shape[0]
    print(f"{name}: {count} records")
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df = df.set_index('Datetime')

# -------------------------
# 2 Define columns
# -------------------------
cols_to_interpolate = ['Indoor Temp','Indoor Hum']         # indoor, small gaps
cols_to_check_large_gaps = ['Outdoor Temp','Outdoor Hum'] # outdoor, may have large gaps

# -------------------------
# 3 Replace zeros with NaN
# -------------------------
df[cols_to_interpolate + cols_to_check_large_gaps] = df[cols_to_interpolate + cols_to_check_large_gaps].replace(0, np.nan)

# -------------------------
# 4 Handle indoor columns (small gaps)
# -------------------------
df[cols_to_interpolate] = df[cols_to_interpolate].interpolate(method='time')

# -------------------------
# 5 Handle outdoor columns (small + large gaps)
# -------------------------
def mark_large_gaps(series, max_gap=50):
    """
    Returns a Series with the size of consecutive NaN gaps.
    """
    is_nan = series.isna()
    gap_sizes = is_nan.groupby((is_nan != is_nan.shift()).cumsum()).transform('sum')
    return gap_sizes

for col in cols_to_check_large_gaps:
    gap_sizes = mark_large_gaps(df[col])
    # Small gaps: interpolate
    df.loc[gap_sizes <= 50, col] = df.loc[gap_sizes <= 50, col].interpolate(method='time')
    # Large gaps: forward fill (you could also use backward fill if needed)
    df.loc[gap_sizes > 50, col] = df.loc[gap_sizes > 50, col].fillna(method='ffill')

# -------------------------
# 6 Check results
# -------------------------
print(df.isna().sum())
```

```python
X = df.drop(['Indoor Temp', 'Indoor Hum'], axis=1)
y = df[['Indoor Temp', 'Indoor Hum']]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)   # numpy array
```

```python
print(np.isnan(X_scaled).sum())
print(np.isnan(y.values).sum())
```

```python
def create_sequences(X, y, window_size=30):
    X_seq = []
    y_seq = []

    for i in range(window_size, len(X)):
        X_seq.append(X[i-window_size:i])  # past 60 rows
        y_seq.append(y[i])                # next row target

    return np.array(X_seq), np.array(y_seq)


# X is your scaled features (numpy array)
# y is your target: Indoor Temp + Indoor Hum (numpy array)

X_lstm, y_lstm = create_sequences(X_scaled, y.values, window_size=30)

print("X_lstm shape:", X_lstm.shape)
print("y_lstm shape:", y_lstm.shape)
```

```python
train_size = int(0.8 * len(X_lstm))  # 80% for training

X_train, X_test = X_lstm[:train_size], X_lstm[train_size:]
y_train, y_test = y_lstm[:train_size], y_lstm[train_size:]

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
model = Sequential()
model.add(LSTM(128, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
model.add(Dropout(0.3))  # increased from 0.2 to 0.3
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.3))  # increased from 0.2 to 0.3
model.add(Dense(32, activation='relu'))
model.add(Dense(y_train.shape[1]))  # output 2 features
optimizer = Adam(learning_rate=0.0005)
model.compile(optimizer= optimizer, loss='mse', metrics=['mae'])
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
```

```python
history = model.fit(
    X_train, y_train,
    epochs=60,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[early_stop],
    verbose=1
)
```

```python
# Take last 30 rows from your X (scaled features)
last_30 = X_scaled[-30:]  # shape (30, 6)

# Reshape for LSTM
test_input = last_30.reshape(1, 30, 6)
```

```python
pred = model.predict(test_input)

predicted_temp = pred[0][0]
predicted_hum  = pred[0][1]

print("Next Indoor Temp Prediction =", predicted_temp)
print("Next Indoor Hum  Prediction =", predicted_hum)
```

```python
future_steps = 6   # 6 steps = next 30 mins
future_preds = []

current_seq = last_30.copy()

for _ in range(future_steps):
    inp = current_seq.reshape(1, 30, 6)
    pred = model.predict(inp)[0]

    future_preds.append(pred)

    # shift window: remove first row, add predicted row? No.
    #  Predicted values are only Indoor Temp & Hum  2 columns.
    # The other 6 inputs are unknown  so keep last known X values.

    next_row = current_seq[-1].copy()
    next_row[:2] = pred  # Replace Indoor Temp, Indoor Hum

    current_seq = np.vstack([current_seq[1:], next_row])
```

```python
for i, p in enumerate(future_preds):
    print(f"Step {i+1}: Temp={p[0]:.2f}, Hum={p[1]:.2f}")
```
